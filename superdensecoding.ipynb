```python
from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister
from qiskit_aer.primitives import Sampler
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram
```

```python
c = '0'
d = '1'
```

```python
ebit0 = QuantumRegister(1,"Alice")
ebit1 = QuantumRegister(1,"Bob")

protocol = QuantumCircuit(ebit0,ebit1)

protocol.h(0)
protocol.cx(0,1)
protocol.barrier()

if d == 1:
    protocol.z(0)
if c == 1:
    protocol.x(0)
protocol.barrier()

protocol.cx(0,1)
protocol.h(0)

protocol.measure_all()

display(protocol.draw(output='mpl'))
```

```python
results = Sampler().run(protocol).result()
stats = results.quasi_dists[0].binary_probabilities()
display(plot_histogram(stats))
```
